$ m = FOO = 0101\ 1110\ 1110 $

**Encryption**

_Encryption algorithm was taken from lecture slide 9 for ease of referencing, please refer to those for more definitions._

$ c_0 = IV $

$ c_i = E_k(m_i \oplus c_{i-1}) $ for $ i = 1,2, ..., N $

$ c = \varepsilon_k(m) = c_0 || c_1 || c_2 || ... || c_N $

1. Split the plaintext into blocks of 4 bits.

$$ m_1 = 0101 $$

$$ m_2 = 1110 $$

$$ m_3 = 1110 $$

2. Initialize the initialization vector $IV$ to a random value with 4 bits. In this case, the question specifies the $IV$ to use.

$$ IV = 1010 $$

3. XOR each block's plaintext, $m_i$, with the previous block's ciphertext, $c_{i-1}$, and then run that through the block cipher, $E_k$, to retrieve ciphertext, $c_i$, for that block. Note that the first plaintext block's "previous ciphertext block" is the initialization vector, $IV$.

$$ c_1 = E_k(0101 \oplus 1010) = E_k(1111) = 1111 $$

$$ c_2 = E_k(1110 \oplus 1111) = E_k(0001) = 0001 $$

$$ c_3 = E_k(1110 \oplus 0001) = E_k(1111) = 1111 $$

4. "Combine" the ciphertext from each block to form the ciphertext for the entire message:

$$ c = 1010\ 1111\ 0001\ 1111 = KPBP $$

# Problem 2 Implementing a Meet-in-the-Middle Attack on a Mini Block Cipher



```python
class SimplifiedAES(object):
    """Simplified AES is a simplified version of AES algorithm"""

    # S-Box
    sBox = [
        0x9,
        0x4,
        0xA,
        0xB,
        0xD,
        0x1,
        0x8,
        0x5,
        0x6,
        0x2,
        0x0,
        0x3,
        0xC,
        0xE,
        0xF,
        0x7,
    ]

    # Inverse S-Box
    sBoxI = [
        0xA,
        0x5,
        0x9,
        0xB,
        0x1,
        0x7,
        0x8,
        0xF,
        0x6,
        0x0,
        0x2,
        0x3,
        0xC,
        0x4,
        0xD,
        0xE,
    ]

    def __init__(self, key):
        # Round keys: K0 = w0 + w1; K1 = w2 + w3; K2 = w4 + w5
        self.pre_round_key, self.round1_key, self.round2_key = self.key_expansion(key)

    def sub_word(self, word):
        """ Substitute word

        :param word: word
        """
        # Take each nibble in the word and substitute another nibble for it using
        # the Sbox table
        return (self.sBox[(word >> 4)] << 4) + self.sBox[word & 0x0F]

    def rot_word(self, word):
        """ Rotate word

        :param word: word
        """
        # Swapping the two nibbles in the word since eqv to rotate here
        return ((word & 0x0F) << 4) + ((word & 0xF0) >> 4)

    def key_expansion(self, key):
        """Key expansion

        Creates three 16-bit round keys from one single 16-bit cipher key

        Cipher Key : | n0 | n1 | n2 | n3 |
        w[0]       : | n0 | n1 |
        w[1]       : | n2 | n3 |

        for i % 2 == 0:
            w[i] : w[i - 2] XOR (SubWord(RotWord(W[i-1])) XOR RC[Nr])
        else:
            w[i] = w[i - 1] XOR w[i - 2]

        :param key: key to be used for encryption and/or decryption
        :returns: Tuple containing pre-round, round 1 and round 2 key in order
        """

        # Round constants
        Rcon1 = 0x80
        Rcon2 = 0x30

        # Calculating value of each word
        w = [None] * 6
        w[0] = (key & 0xFF00) >> 8
        w[1] = key & 0x00FF
        w[2] = w[0] ^ (self.sub_word(self.rot_word(w[1])) ^ Rcon1)
        w[3] = w[2] ^ w[1]
        w[4] = w[2] ^ (self.sub_word(self.rot_word(w[3])) ^ Rcon2)
        w[5] = w[4] ^ w[3]

        return (
            self.int_to_state((w[0] << 8) + w[1]),  # Pre-Round key
            self.int_to_state((w[2] << 8) + w[3]),  # Round 1 key
            self.int_to_state((w[4] << 8) + w[5]),  # Round 2 key
        )

    def gf_mult(self, a, b):
        """Galois field multiplication of a and b in GF(2^4) / x^4 + x + 1
        :param a: First number
        :param b: Second number
        :returns: Multiplication of both under GF(2^4)
        """
        # Initialise
        product = 0

        # Mask the unwanted bits
        a = a & 0x0F
        b = b & 0x0F

        # While both multiplicands are non-zero
        while a and b:

            # If LSB of b is 1
            if b & 1:

                # Add current a to product
                product = product ^ a

            # Update a to a * 2
            a = a << 1

            # If a overflows beyond 4th bit
            if a & (1 << 4):

                # XOR with irreducible polynomial with high term eliminated
                a = a ^ 0b10011

            # Update b to b // 2
            b = b >> 1

        return product

    def int_to_state(self, n):
        """Convert a 2-byte integer into a 4-element vector (state matrix)
        :param m: integer
        :returns: state corresponding to the integer
        """
        return [n >> 12 & 0xF, (n >> 4) & 0xF, (n >> 8) & 0xF, n & 0xF]

    def state_to_int(self, m):
        """Convert a 4-element vector (state matrix) into 2-byte integer
        :param m: state
        :returns: integer corresponding to the state
        """
        return (m[0] << 12) + (m[2] << 8) + (m[1] << 4) + m[3]

    def add_round_key(self, s1, s2):
        """Add round keys in GF(2^4)

        :param s1: First number
        :param s2: Second number
        :returns: Addition of both under GF(2^4)
        """
        return [i ^ j for i, j in zip(s1, s2)]

    def sub_nibbles(self, sbox, state):
        """Nibble substitution

        :param sbox: Substitution box to use for transformatin
        :param state: State to perform sub nibbles transformation on
        :returns: Resultant state
        """
        return [sbox[nibble] for nibble in state]

    def shift_rows(self, state):
        """Shift rows and inverse shift rows of state matrix (same)

        :param state: State to perform shift rows transformation on
        :returns: Resultant state
        """
        return [state[0], state[1], state[3], state[2]]

    def mix_columns(self, state):
        """Mix columns transformation on state matrix

        :param state: State to perform mix columns transformation on
        :returns: Resultant state
        """
        return [
            state[0] ^ self.gf_mult(4, state[2]),
            state[1] ^ self.gf_mult(4, state[3]),
            state[2] ^ self.gf_mult(4, state[0]),
            state[3] ^ self.gf_mult(4, state[1]),
        ]

    def inverse_mix_columns(self, state):
        """Inverse mix columns transformation on state matrix

        :param state: State to perform inverse mix columns transformation on
        :returns: Resultant state
        """
        return [
            self.gf_mult(9, state[0]) ^ self.gf_mult(2, state[2]),
            self.gf_mult(9, state[1]) ^ self.gf_mult(2, state[3]),
            self.gf_mult(9, state[2]) ^ self.gf_mult(2, state[0]),
            self.gf_mult(9, state[3]) ^ self.gf_mult(2, state[1]),
        ]

    def encrypt(self, plaintext):
        """Encrypt plaintext with given key

        Example::

            ciphertext = SimplifiedAES(key=0b0100101011110101).encrypt(0b1101011100101000)

        :param plaintext: 16 bit plaintext
        :returns: 16 bit ciphertext
        """
        state = self.add_round_key(self.pre_round_key, self.int_to_state(plaintext))

        state = self.mix_columns(self.shift_rows(self.sub_nibbles(self.sBox, state)))

        state = self.add_round_key(self.round1_key, state)

        state = self.shift_rows(self.sub_nibbles(self.sBox, state))

        state = self.add_round_key(self.round2_key, state)

        return self.state_to_int(state)

    def decrypt(self, ciphertext):
        """Decrypt ciphertext with given key

        Example::

            plaintext = SimplifiedAES(key=0b0100101011110101).decrypt(0b0010010011101100)

        :param ciphertext: 16 bit ciphertext
        :returns: 16 bit plaintext
        """
        state = self.add_round_key(self.round2_key, self.int_to_state(ciphertext))

        state = self.sub_nibbles(self.sBoxI, self.shift_rows(state))

        state = self.inverse_mix_columns(self.add_round_key(self.round1_key, state))

        state = self.sub_nibbles(self.sBoxI, self.shift_rows(state))

        state = self.add_round_key(self.pre_round_key, state)

        return self.state_to_int(state)
```

To implement the meet-in-the-middle (MitM) attack strategy on the mini block cipher, the idea is to find a matching intermediate state X in both encryption and decryption processes to recover the key pair (Key1, Key2). Here's how we can approach the attack:
Meet-in-the-Middle Attack Strategy

    Encrypt Partially:
        For a given plaintext P, perform only the first round of encryption (using Key1) to get the intermediate state X.

    Decrypt Partially:
        For a given ciphertext C, perform only the second round of decryption (using Key2) to get an intermediate state X'.

    Matching Keys:
        Compare the intermediate states X and X'. If they match, store the corresponding key pair (Key1, Key2).
        There could be multiple matching key pairs, so repeat with another plaintext-ciphertext pair to narrow down to the correct key pair.

Steps to Implement:

    Generate Keys:
        Assume we have multiple candidates for Key1 and Key2 (e.g., brute-force all possible 16-bit keys).

    Partial Encryption & Decryption:
        Implement functions for partial encryption (up to X) and partial decryption (from X').

    Meet-in-the-Middle Attack:
        For each possible Key1, compute the intermediate state X for a given plaintext.
        For each possible Key2, compute the intermediate state X' for the corresponding ciphertext.
        Check if X == X'.

    Narrow Down the Key:
        Use another plaintext-ciphertext pair to eliminate incorrect key pairs.

```python
class SimplifiedAESMITM:
    def __init__(self, known_plaintext, known_ciphertext):
        self.known_plaintext = known_plaintext
        self.known_ciphertext = known_ciphertext

    def mitm_attack(self):
        # Dictionary to store intermediate results with key K1
        middle_text_dict = {}

        # Generate and store intermediate results for every possible K1
        for K1 in range(0x10000):  # Assuming 16-bit key space
            cipher1 = SimplifiedAES(K1)
            middle_text = cipher1.encrypt(self.known_plaintext)
            middle_text_dict[middle_text] = K1

        # Try every possible K2 and see if it matches with an intermediate result
        for K2 in range(0x10000):
            cipher2 = SimplifiedAES(K2)
            decrypted_middle_text = cipher2.decrypt(self.known_ciphertext)

            # Check if this decrypted middle text matches any stored middle text
            if decrypted_middle_text in middle_text_dict:
                K1 = middle_text_dict[decrypted_middle_text]
                print(f"Found key pair: K1 = {format(K1, '016b')}, K2 = {format(K2, '016b')}")
                return K1, K2

        print("No matching keys found.")
        return None, None

# Define known plaintext and ciphertext
known_plaintext = 0b0101101001000011
known_ciphertext = 0b1010100111110110

# Instantiate and perform the MITM attack
mitm = SimplifiedAESMITM(known_plaintext, known_ciphertext)
K1, K2 = mitm.mitm_attack()

if K1 is not None and K2 is not None:
    print(f"Successful MITM attack! Key pair found: K1 = {format(K1, '016b')}, K2 = {format(K2, '016b')}")
else:
    print("MITM attack failed. No key pair found.")
```

```python
import random

# SimplifiedAES class definition should go here (your provided code)

class SimplifiedAESMITM:
    def __init__(self, known_plaintext, known_ciphertext):
        self.known_plaintext = known_plaintext
        self.known_ciphertext = known_ciphertext

    def mitm_attack(self):
        # Dictionary to store intermediate results with key K1
        middle_text_dict = {}

        # Generate and store intermediate results for every possible K1
        for K1 in range(0x10000):  # Assuming 16-bit key space
            cipher1 = SimplifiedAES(K1)
            middle_text = cipher1.encrypt(self.known_plaintext)
            middle_text_dict[middle_text] = K1

        # Try every possible K2 and see if it matches with an intermediate result
        for K2 in range(0x10000):
            cipher2 = SimplifiedAES(K2)
            decrypted_middle_text = cipher2.decrypt(self.known_ciphertext)

            # Check if this decrypted middle text matches any stored middle text
            if decrypted_middle_text in middle_text_dict:
                K1 = middle_text_dict[decrypted_middle_text]
                return K1, K2  # Return the successful key pair

        return None, None  # Return None if no matching keys are found

# Define the key
key = 0b0101100101111010
cipher = SimplifiedAES(key)

# Generate 10 random plaintext-ciphertext pairs
plaintext_ciphertext_pairs = []
for _ in range(10):
    plaintext = random.randint(0, 0xFFFF)  # Generate random 16-bit plaintext
    ciphertext = cipher.encrypt(plaintext)  # Encrypt to get the ciphertext
    plaintext_ciphertext_pairs.append((plaintext, ciphertext))

# Loop through each plaintext-ciphertext pair and perform the MITM attack
for index, (known_plaintext, known_ciphertext) in enumerate(plaintext_ciphertext_pairs, start=1):
    print(f"\nPerforming MITM attack on Pair {index}:")
    print(f"Plaintext: {format(known_plaintext, '016b')} | Ciphertext: {format(known_ciphertext, '016b')}")

    # Instantiate MITM attack for the current pair
    mitm = SimplifiedAESMITM(known_plaintext, known_ciphertext)
    K1, K2 = mitm.mitm_attack()

    if K1 is not None and K2 is not None:
        print(f"Successful MITM attack! Key pair found: K1 = {format(K1, '016b')}, K2 = {format(K2, '016b')}")
    else:
        print("MITM attack failed. No key pair found.")
```

## **Task 3: Analyze the time and memory complexity of the attack compared with the naive exhaustive key search.**

*a) What is the key space for the mini block cipher?*

To identify the keyspace the mini block cipher uses an n-bit key. The keyspace is then 2^n. Where n represents the bit size

For a mini block cipher that is 16 bits the key space would be: 2^16. This means that there are 65,536 possible keys.

b) *Imagine the mini block cipher is executed twice to generate a cipher text. It is called a double mini cipher block. We need a key in 32 bits. The first 16 to the first mini block cipher, the remaining 16 to the second mini block cipher. The meet in the middle attack is to match the state for the first encryption of mini block cipher and the second decryption mini block. How many operations are needed for such an attack?*

Double mini block ciphers would divide the 32 bits into two 16 bit keys since the double mini block cipher uses two keys. Each key will be 16 bits long.
The key space would be 2^16 x 2^16. Making 2^32. The double mini block cipher would have 4,294,967,296 possible keys.

In order to perform a meet-in-the-middle attack first you need to encrypt the plain text. This will be done using the first mini cipher at 2^16 and store each resulting intermediate state.. following the encryption phase you will now be in the decryption phase. In order to decrypt the ciphertext you must use the second mini ciphers with 2^16 possible keys and store each resulting intermediate state.. During this phase you will also be matching the intermediate states to find any and all matches. this will reveal the correct pair of keys.

The encryption phase performed 2^16. The decryption phase performed 2^16. when adding this together the total operations is : 2^16 + 2^16 = 2^17. The meet-in-the-middle attack required 2^17 or 131,072 Operations. This attack is efficient because it only requires storing a table of size 2^16,  which is significantly smaller compared to the full key space of 2^32.

c) *If we do exhaustive key search for the double mini block cipher, how many operations are needed?*

An exhaustive key search for the double mini block cipher, a 32-bit key is typically used. This is split into the two 16 bit keys. Each key will be 16 bits long.
The key space would be 2^16 x 2^16. Making 2^32. The double mini block cipher would have 4,294,967,296 possible keys. With exhaustive key search  the requirement is to test each possible key in the 32-bit combination. This means we would have to perform 2^32 operations.

d) *What is the tradeoff for the MITM attack (speed, memory, etc.)*

The meet-in-the-middle attack trades off memory for a smaller amount of operations.
Exhaustive key search requires 2^32  possible keys for the 32 bit key. The memory for the tests are minimal as each key is tested without storing in the intermediate states.
The meet-in-the-middle attack  requires 2^17 operations for the 2^32 key function as it is achieved by only doing 2^17 operations. The memory for the operations is significantly more than exhaustive key search because the 2^16 operations are stored for encryption as well as 2^16th intermediate decryption states. This means that 2^17 entries are saved in the memory allocation.

When comparing meet-in-the-middle attacks and exhaustive key search the real decision on which to use really matters on whether or not you have the time or space. If memory is an issue and you have a limited space for memory the recommendation would be exhaustive key search as the test for the keys are not saved. Though this may take a bit more time you are able to save on memory space. If time is an issue and you have the available memory allocation, a meet-in-the-middle attack would be the better option as it takes less time due to the fact that it has less operations and stores more in the memory.

Overall meet-in-the-middle attack is more used when memory is available as it reduces the operations and workload when compared to exhaustive key search. However if memory is a concern exhaustive key search may be the only option.
